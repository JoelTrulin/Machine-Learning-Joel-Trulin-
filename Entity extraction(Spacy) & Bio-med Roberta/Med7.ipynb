```python
import pandas as pd
import spacy
med7 = spacy.load("en_core_med7_lg")
nlp = spacy.load('en_core_web_sm')
```

Load Data

```python
data = pd.read_excel('/content/drive/MyDrive/med7data.xlsx')
data['preproc'] = data.methods.apply(lambda x: list(nlp(x).sents))
data = data.drop('methods', axis=1)
df = data.explode('preproc').rename(columns={'preproc':'token_sentc'})
df['entities'] = df.token_sentc.apply(lambda x: [{ent.text: ent.label_} for ent in med7(str(x)).ents])
```

```python
data = data.join(df)
#data.to_excel('/content/drive/MyDrive/Data with Entities.xlsx')
data.to_csv('/content/drive/MyDrive/Data with Entities.csv')
```
